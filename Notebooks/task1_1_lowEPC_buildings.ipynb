# General Functions

```python
col_keep = [
    'LMK_KEY', 'UPRN', 'ADDRESS1', 'POSTCODE',
    'BUILDING_REFERENCE_NUMBER', 'CURRENT_ENERGY_RATING', 'CURRENT_ENERGY_EFFICIENCY',
    'PROPERTY_TYPE', 'BUILT_FORM', 'INSPECTION_DATE', 'ENERGY_CONSUMPTION_CURRENT',
    'CO2_EMISS_CURR_PER_FLOOR_AREA', 'TOTAL_FLOOR_AREA',
    'MAIN_FUEL', 'CONSTRUCTION_AGE_BAND',
]

low_ratings = ['E', 'F', 'G']
```

```python
def process_epc_dataframe(df, district):
    import numpy as np
    df = df.copy()

    # Transferring date format
    df['INSPECTION_DATE'] = pd.to_datetime(df['INSPECTION_DATE'], errors='coerce')

    # Filtering time range
    df = df[
        (df['INSPECTION_DATE'] >= '2022-01-01') & (df['INSPECTION_DATE'] <= '2024-12-31')
    ]
    
    # Grouping data by 'BUILDING_REFERENCE_NUMBER' for newest recording
    df = df.sort_values('INSPECTION_DATE').groupby('BUILDING_REFERENCE_NUMBER', as_index=False).last()
    
    # Keeping the target col
    existing_cols = [col for col in col_keep if col in df.columns]
    df = df[existing_cols]
    
    # Adding district
    df['District'] = district
    
    # Filtering low EPC ratings
    low_ratings = ['E', 'F', 'G']
    df = df[df['CURRENT_ENERGY_RATING'].isin(low_ratings)]
    
    # Imputing missing values
    for col in df.columns:
        if pd.api.types.is_datetime64_any_dtype(df[col]):
            continue
        elif df[col].dtype == 'object':
            df[col] = df[col].fillna('Unknown')
        elif np.issubdtype(df[col].dtype, np.number):
            df[col] = df[col].fillna(-1)

    # Summarizing results
    print(f'\nMissing values in {district}:')
    print(df.isna().sum())

    return df
```

```python
district_files = {
    'Bolton': df_bolton,
    'Bury': df_bury,
    'Manchester': df_manchester,
    'Oldham': df_oldham,
    'Rochdale': df_rochdale,
    'Salford': df_salford,
    'Stockport': df_stockport,
    'Tameside': df_tameside,
    'Trafford': df_trafford,
    'Wigan': df_wigan
}

processed_dfs = {}

for name, df in district_files.items():
    processed_dfs[name] = process_epc_dataframe(df, name)
```

# Merging postcodes

```python
df_postcode.rename(columns={'postcode': 'POSTCODE'}, inplace=True)
def add_geolocation(df, postcode_df):
    
    # Geolocation
    df = df.merge(postcode_df[['POSTCODE', 'latitude', 'longitude']], on='POSTCODE', how='left')

    missing_geo = df['latitude'].isna().sum()
    missing_address = df['ADDRESS1'].isna().sum()

    print(f'Postcode geolocation missing: {missing_geo}')
    print(f'Missing ADDRESS1: {missing_address}')

    df = df.dropna(subset=['latitude', 'longitude', 'ADDRESS1'])

    return df
```

```python
for name, df in processed_dfs.items():
    print(f'\nAdding geolocation to {name}')
    processed_dfs[name] = add_geolocation(df, df_postcode)
```

```python
for name, df in processed_dfs.items():
    df.to_csv(f'{name}_processed.csv', index=False)
    print(f'Exported: {name}_processed.csv')
```
